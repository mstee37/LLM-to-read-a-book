```python
import pandas as pd
```

```python
logic_ans = pd.read_csv("Logic_Answers.csv")
logic_qn = pd.read_csv("Logic_Questions.csv")
stats_ans = pd.read_csv("Statistics_Answers.csv")
stats_qn = pd.read_csv("Statistics_Questions.csv")
```

```python
logic_ans["Number"] = logic_ans["Item Number"].str.extract(r"(\d+\.\d+)", expand=False)
logic_qn["Number"] = logic_qn["Question Number"].str.extract(r"(\d+\.\d+)", expand=False)
```

```python
logic_merge = logic_qn.merge(logic_ans, on="Number", how="outer")
logic_merge.rename(columns={"Content_x":"Question","Content_y":"Answer", "Item Number":"Answer Number"}, inplace=True)
```

```python
logic_merge.to_csv("Combined_Logic.csv")
logic_merge
```

```python
stats_ans["Number"] = stats_ans["Item Number"].str.extract(r"(\d+\.\d+)", expand=False)
stats_qn["Number"] = stats_qn["Question Number"].str.extract(r"(\d+\.\d+)", expand=False)
stats_merge = stats_qn.merge(stats_ans, on="Number", how="outer")
stats_merge.rename(columns={"Content_x":"Question","Content_y":"Answer", "Item Number":"Answer Number"}, inplace=True)
stats_merge.isna().sum()
```

```python
stats_merge.to_csv("Combined_Stats.csv")
stats_merge
```
