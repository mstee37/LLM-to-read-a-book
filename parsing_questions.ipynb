```python
from pdf_parsing import extract_questions_to_dataframe
import re
import pandas as pd
```

```python
pdf_path = "heard-on-the-street-quantitative-questions-from-wall-street-job-interviews-revised-22nd.pdf"  # Replace with the path to your PDF
pages_to_extract = [x for x in range(90,100)]  # Replace with the specific pages you need

# Extract questions into DataFrame
questions_df = extract_questions_to_dataframe(pdf_path, pages_to_extract)
```

```python
questions_df
```

```python
# Function to parse the questions into a DataFrame
def parse_questions(text):
    # Regex pattern to match questions
    pattern = r"(Question\s+\d+\.\d+:.*?)(?=Question\s+\d+\.\d+:|$)"
    matches = re.findall(pattern, text, re.DOTALL)
    
    # Process each match to extract question number and content
    questions_data = []
    for match in matches:
        # Extract the question number and content
        question_number_match = re.search(r"(Question\s+\d+\.\d+:)", match)
        question_number = question_number_match.group(0) if question_number_match else ""
        question_content = match.replace(question_number, "").strip()
        
        # Add to list
        questions_data.append({"Question Number": question_number.strip(":"), "Content": question_content})
    
    # Convert to DataFrame
    df = pd.DataFrame(questions_data)
    return df
```

```python
df = pd.DataFrame()
```

```python
df = pd.DataFrame()

for i in questions_df.Content:
    # print(i)
    df = pd.concat([df, parse_questions(i)], axis=0)

df = df.reset_index(drop=True)
```
