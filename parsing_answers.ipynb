```python
def parse_items(text, pattern=r"(Answer\s+\d+\.\d+:.*?)(?=Answer\s+\d+\.\d+:|$)"):
    """
    Parse items (e.g., questions or answers) from a given text based on a provided regex pattern.

    Args:
        text (str): The text content to parse items from.
        pattern (str): The regex pattern to identify and split items.

    Returns:
        pd.DataFrame: A DataFrame containing item numbers and their content.
    """
    # Find all matches based on the pattern
    matches = re.findall(pattern, text, re.DOTALL)
    
    # Process each match to extract item number and content
    items_data = []
    for match in matches:
        # Extract the item number and content dynamically
        item_number_match = re.search(r"(Answer\s+\d+\.\d+:)", match)
        item_number = item_number_match.group(0) if item_number_match else ""
        item_content = match.replace(item_number, "").strip()
        
        # Add to list
        items_data.append({"Item Number": item_number.strip(":"), "Content": item_content})
    
    # Convert to DataFrame
    df = pd.DataFrame(items_data)
    return df
```

```python
df = pd.DataFrame()
```

```python
df = pd.DataFrame()

for i in answer_df.Content:
    # print(i)
    df = pd.concat([df, parse_items(i)], axis=0)

df = df.reset_index(drop=True)
```
